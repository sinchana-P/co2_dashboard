# (0) Some minor data preprocessing

```python
# Fill NAs with 0s and create GDP per capita column
df = df.fillna(0)
df['gdp_per_capita'] = np.where(df['population']!= 0, df['gdp']/ df['population'], 0 )
```

```python
df['gdp_per_capita'].tail(20)
```

```python
# Make a DataFrame Pipeline Interactive
idf = df.interactive()
```

# (1) CO2 EMISSION OVER TIME BY CONTINENT

# work flow
1. get year_slider
2. get YEAR input --> via year_slider
3. Build y-axis values  -->  yaxis_co2   ---> via RadioButton
4. create "PIPELINE" --->  ( make dataframe interactive with widgets )
5. CONNECT "DATA PIPELINE" TO "WIDGET"

```python
# Define PANEL WIDGETS
# PANEL WIDGETS

year_slider = pn.widgets.IntSlider( name = 'Year Slider', start = 1750, end = 2020, step = 5, value = 1850)
year_slider
```

```python
# Radion buttons for CO2 measures
# PANEL WIDGETS

yaxis_co2 = pn.widgets.RadioButtonGroup(
    name = 'Y axis',
    options = ['co2', 'co2_per_capita',],
    button_type = 'success'
)
```

```python
yaxis_co2
```

```python
# use idf

continents = ['World', 'Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'Antartica']

co2_pipeline = (
    idf[
        ( idf.year <= year_slider ) &
        ( idf.country.isin(continents) )
    ]
    .groupby(['country', 'year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop = True)
)
```

```python
df.columns
```

```python
co2_pipeline
```

## CONNECT "DATA PIPELINE" TO "WIDGET"

```python
# final plot
# define : x - axis, y - axis
# use PIPELINE : co2_pipeline

co2_plot = co2_pipeline.hvplot(x = 'year', by = 'country', y = yaxis_co2, line_width = 2, title = "CO2 emission by continent")
co2_plot
```

# (2) Table - CO2 emission over time by continent

```python
co2_table = co2_pipeline.pipe( pn.widgets.Tabulator, pagination = 'remote', page_size = 10, sizing_mode = 'stretch_width' )
co2_table
```

# (3) CO2 vs GDP scatterplot

```python
# create different pipeline for 3rd visualization (i.e. scatter)

co2_vs_gdp_scatterplot_pipeline = (
    idf[
        ( idf.year == year_slider ) &
        ( ~ ( idf.country.isin(continents) ) )
    ]
    .groupby([ 'country', 'year', 'gdp_per_capita' ])['co2'].mean()
    .to_frame()
    .reset_index()
    .sort_values(by = 'year')
    .reset_index(drop = True)
)
```

```python
co2_vs_gdp_scatterplot_pipeline
```

```python
co2_vs_gdp_scatterplot = co2_vs_gdp_scatterplot_pipeline.hvplot(x = 'gdp_per_capita',
                                                                y = 'co2',
                                                                by = 'country',
                                                                size = 80,
                                                                kind = "scatter",
                                                                alpha = 0.7,
                                                                legend = False,
                                                                height = 500,
                                                                width = 500
                                                               )
```

```python
co2_vs_gdp_scatterplot
```

# (4) Bar chart with CO2 sources by continent

```python
yaxis_co2_source = pn.widgets.RadioButtonGroup(name = 'Y axis',
                                              options = ['coal_co2', 'oil_co2', 'gas_co2'],
                                              button_type = 'success'
                                             )
```

```python
continents_excl_world = ['Asia', 'Oceania', 'Europe', 'Africa', 'North America', 'Antarctica']

co2_source_bar_pipeline = (
    idf[
        ( idf.year == year_slider ) & 
        ( idf.country.isin(continents_excl_world) )
    ]
    .groupby(['year', 'country'])[yaxis_co2_source].sum()
    .to_frame()
    .reset_index()
    .sort_values(by = 'year')
    .reset_index(drop = True)
)
```

```python
co2_source_bar_plot = co2_source_bar_pipeline.hvplot(kind = 'bar',
                                                     x = 'country',
                                                     y = yaxis_co2_source,
                                                     title = 'CO2 source by continent'
                                                    )
```

```python
co2_source_bar_plot
```
